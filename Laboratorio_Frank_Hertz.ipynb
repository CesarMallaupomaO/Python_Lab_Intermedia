```python
import pandas as pd
import numpy as np


# Ruta del archivo con los datos del experimento
archivo_excel = r"C:\Users\César\OneDrive\Documentos\Laboratorios\Neon_dia2_grupo4.xlsx"
# Cargar el archivo Excel
excel_file = pd.ExcelFile(archivo_excel)

# Obtener las hojas del archivo Excel
# sheet_names : Es un método que devuelve una lista con 
# los nombres de todas las hojas en el archivo Excel.
hojas = excel_file.sheet_names

#Declarando e inicializando
diferencias1 = np.empty(51)
diferencias2 = np.empty(51)
diferencias3 = np.empty(51)
difer4 = np.empty(0)
indice = 0
##########################Analisis_por_minimos#################################
for hoja in hojas:
    # Leer datos de la hoja actual  
    datos_hoja = excel_file.parse(hoja)
      
    # Seleccionar los datos de la columna  "U1[V]'" que tienen el tipo " minimo"
    datos_maximos = datos_hoja[datos_hoja['tipo'] == 'máximo']['U1[V]']
    columna = np.array(datos_maximos)

   #Se estan restando los minimos consecutivos
    dif1 = columna[1] - columna[0] 
    diferencias1[indice] = dif1
    dif2 = columna[2] - columna[1] 
    diferencias2[indice] = dif2
    dif3 = columna[3] - columna[2] 
    diferencias3[indice] = dif3
    
    #Algunos de los datos tienen 5 datos
    if  len(columna) > 4:
        # print(" columna[3]  : {:.2f} ".format(columna[3]))
        # print(" columna[4]  : {:.2f} ".format(columna[4]))
        dif4 = columna[4] - columna[3] 
        difer4 = np.append(difer4, dif4)
        
    # variable contadora
    indice = indice + 1
diferencias4 = np.delete(difer4 , 0)  

#calculando la media y su desviacion estandar, 
media1 = np.mean(diferencias1)
media2 = np.mean(diferencias2)
media3 = np.mean(diferencias3)
media4 = np.mean(diferencias4)
desviacion_estandar1 = np.std(diferencias1)
desviacion_estandar2 = np.std(diferencias2)
desviacion_estandar3 = np.std(diferencias3)
desviacion_estandar4 = np.std(diferencias4)
#imprimir en consola los valores obtenidos
print("El primer valor medio: {:.2f}  su desviación estandar: {:.2f}".format(media1, desviacion_estandar1))
print("El segundo valor medio: {:.2f} su desviacion estandar: {:.2f}".format(media2, desviacion_estandar2))
print("El tercer valor medio: {:.2f}  su desviacion estandar: {:.2f}".format(media3, desviacion_estandar3))
print("El cuarto valor medio: {:.2f}  su desviacion estandar: {:.2f}".format(media4, desviacion_estandar4))

########################################################################
```

```python
#%% Grafica Maximos

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Ruta del archivo Excel
archivo_excel = r"C:\Users\César\Desktop\Imagenes_Hertz.xlsx"

# Cargar el archivo Excel
excel_img = pd.ExcelFile(archivo_excel)

# Leer datos de la hoja
datos_hoja0 = excel_img.parse("Hoja3")
x1 = datos_hoja0['Voltage U1']
y1 = datos_hoja0['Corriente IA']

# Crear el gráfico y superponer las curvas
plt.plot(x1, y1, label='U2: 7.0 V', color='black')

# Agregar etiquetas y leyenda
plt.xlabel(r'$\mathrm{Voltage\ U_1\ [V]}$')
plt.ylabel(r'$\mathrm{Corriente\ I\ [nA]}$')
plt.title(r'$\mathrm{U_1\ vs\ I}$')
plt.legend()  # Mostrar leyenda

#########################################################################

pos1 = x1.tolist().index(25)
pos2 = x1.tolist().index(35.01)

pos3 = x1.tolist().index(40.02)
pos4 = x1.tolist().index(55.01)

pos5 = x1.tolist().index(60.02)
pos6 = x1.tolist().index(75.01)

pos7 = x1.tolist().index(80.02)
pos8 = x1.tolist().index(95.01)

valores_intervalo1 = y1[pos1:pos2+1]  # +1 en b para incluir b en el intervalo
valores_intervalo2 = y1[pos3:pos4+1]  
valores_intervalo3 = y1[pos5:pos6+1]  
valores_intervalo4 = y1[pos7:pos8+1] 

 
# Encontrar el máximo valor en el intervalo
max_valory_intervalo1 = max(valores_intervalo1)
max_valory_intervalo2 = max(valores_intervalo2)
max_valory_intervalo3 = max(valores_intervalo3)
max_valory_intervalo4 = max(valores_intervalo4)

# Hallar la posicion de estos valores 
posf1 = y1.tolist().index(max_valory_intervalo1)
posf2 = y1.tolist().index(max_valory_intervalo2)
posf3 = y1.tolist().index(max_valory_intervalo3)
posf4 = y1.tolist().index(max_valory_intervalo4)

valor_x_en_max1 = x1[posf1]
valor_x_en_max2 = x1[posf2]
valor_x_en_max3 = x1[posf3]
valor_x_en_max4 = x1[posf4]


# Coordenadas y valores de los máximos locales
maximos_locales = [(valor_x_en_max1, max_valory_intervalo1), 
                   (valor_x_en_max2, max_valory_intervalo2), 
                   (valor_x_en_max3, max_valory_intervalo3), 
                   (valor_x_en_max4, max_valory_intervalo4)]


##########################################################################


# Agregar flechas y etiquetas para los máximos locales


for x, y in maximos_locales:
    plt.arrow(x, y, 0, 2, color='blue', width=0.2, head_width=2, head_length=1)
    plt.text(x, y + 4, f'Máx: {y} nA', ha='center')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
