```python
#%% GRAFICA AJUSTADA DE LA FUNCION DE CALIBRACIÓN

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

##############################DATOS########################################
Ih = np.array([0.049, 0.252, 0.5, 0.752, 1, 1.253, 1.506, 1.749, 2.08])
Ih_error = np.array([(0.003 + (1 / 100) * t) for t in Ih])

###########################################################################
hBH = np.array([0, 0.1, 0.3, 0.5, 0.7, 0.9, 1, 1.2, 1.4])
hBH_error = np.array([(2 / 100) * j for j in hBH])
################################3##########################################

###Define la función para ajustar####################
def func(Ih, a, b):
    return a * Ih + b
######################################################

####################################AJUSTE###################################
# Realiza el ajuste de curva teniendo en cuenta los errores en x e y, uniendo las 
# dos con una normalización guardada en la variable "sigma"

params, covariance = curve_fit(func, Ih, hBH, p0=[1, 0], sigma=np.sqrt(Ih_error**2 + hBH_error**2), absolute_sigma=True)

# Calcula el error estándar de los parámetros ajustados
param_errors = np.sqrt(np.diag(covariance))

######################DESCRICION_QUE_LLEVA_LA_GRAFICA######################
plt.errorbar(Ih, hBH, yerr=hBH_error, xerr=Ih_error, fmt='o', label='Datos con errores')
Ih_fit = np.linspace(Ih.min(), Ih.max(), 100)
hBH_fit = func(Ih_fit, *params)
plt.plot(Ih_fit, hBH_fit, label='Curva ajustada', color='black')
plt.title('Función de calibración')
plt.xlabel('I(A)')
plt.ylabel('B(mT)')

# Agrega la ecuación a la imagen
a, b = params
a_error, b_error = param_errors
equation = rf'$B = ({a:.4f} \pm {a_error:.4f}) \cdot I + ({b:.4f} \pm {b_error:.4f})$'
plt.text(0.1, 1.2, equation, fontsize=12)

# Muestra el gráfico
plt.legend()
plt.show()


##########################33IMPRESION DE RESULTADOS########################
# Imprime los parámetros ajustados y los errores en los parámetros
print("Ecuación ajustada:", equation)
print("Parámetros ajustados (a, b):", params)
print("Errores en los parámetros (a, b):", param_errors)
```
